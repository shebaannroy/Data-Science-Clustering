<h1>Task 1 – Data Loading and data exploration</h1>

```python
import warnings
warnings. filterwarnings('ignore')
%matplotlib inline
import pandas as pd
import seaborn as sns
import random
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from IPython.core.interactiveshell import InteractiveShell #For displaying all the output for a sigle cell
InteractiveShell.ast_node_interactivity = "all"
from IPython.display import Markdown, display
def printmd(string): #To use Markup language in a code cell
    display(Markdown(string))
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
#from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt # plotting
import numpy as np # linear algebra
import os # accessing directory structure
###########################################################################################################################

printmd("<h3> Reading and loading the data\n")
df=pd.read_csv('customers.csv')
df
printmd("<h3> Columns \n")
list(df.columns)


printmd("<h2>Checking for any null values\n")
df.isnull().sum()


printmd("<h2>Exploring the data\n")
df.info()
print(df.describe())

dfnew=df.copy(deep=True) #Making a copy
###########################################################################################################################

#Performing Data Exploration
printmd("<h1>DATA EXPLORATION")
printmd("<h3>Column types")
print(df.dtypes)

printmd("<h3>Exploring each of the columns")  #Column 1: CustomerID
printmd("<h3>Column: CustomerID")
df['CustomerID'].describe()

printmd("<h3>Column: Gender")#Column 2: Gender
df['Gender'].describe()



sns.set_theme(style="whitegrid")
ax = sns.countplot(x="Gender", data=df)
for i in ax.containers:
    ax.bar_label(i,)
ax.set_title("Bar graph for Gender")
plt.show()

plt.pie(df['Gender'].value_counts(),autopct='%1.1f%%')
plt.show()


printmd("<h3>Column: Age")#Column 3: Age
df['Age'].describe()




sns.histplot(data=df,x='Age',kde=True,color='green').set(title='Histogram for Age')
plt.grid()
sns.set(rc = {'figure.figsize':(10,8)})
plt.show()


printmd("<h3>Column: Annual Income (k$)")#Column 4: Annual Income (k$)
df['Annual Income (k$)'].describe()



sns.histplot(data=df,x='Annual Income (k$)',kde=True).set(title='Histogram for Annual Income')
plt.grid()
sns.set(rc = {'figure.figsize':(10,8)})
plt.show()

printmd("<h3>Column: Spending Score (1-100)")#Column 5: Spending Score (1-100)
df['Spending Score (1-100)'].describe()

sns.histplot(data=df,x='Spending Score (1-100)',kde=True,color='purple').set(title='Histogram for Spending score')
plt.grid()
sns.set(rc = {'figure.figsize':(10,8)})
plt.show()

printmd("<h3>Adding a new column age_group, which categorizes the ages into three classes")#add a new column category next to the age group.
category = pd.cut(df.Age, bins=[18,25,49,70], labels=[ 'young adult', 'middle age', 'senior' ])
df.insert(2,'age_group' ,category)
df

ax = sns.countplot(x="age_group", hue="Gender", data=df).set(title='plot for age group')
sns.set(rc = {'figure.figsize':(10,5)})
plt.show()





#plot for age group vs. annual income
ax = sns.barplot(x="age_group",y="Annual Income (k$)", hue="Gender", data=df).set(title='plot for age group vs. annual income')
plt.show()

#plot for age group vs.Spending Score
ax = sns.barplot(x="age_group",y="Spending Score (1-100)", hue="Gender", data=df).set(title='plot for age group vs. spending score')
plt.show()

#Box plot for age group vs.Spending Score
ax = sns.boxplot(x="age_group",y="Spending Score (1-100)", data=df).set(title='Box plot for age group vs. spending score')
plt.show()
###########################################################################################################################

printmd("<h3>Correlation\n")
del df['CustomerID']
sns.set_theme(style="ticks")
s=sns.pairplot(df, hue="age_group")
plt.show()

#Finding Correlation
printmd("<h3>Correlation Table\n")
df.corr()

printmd("<h3>Correlation Matrix\n")
sns.heatmap(df.corr(), annot=True);
plt.title("Correlation Matrix")
plt.show()

printmd("<h3>Correlation between Age and Spending Score \n")
corr1 = df['Age'].corr(df['Spending Score (1-100)'])
print(corr1)  

scat1=df.plot.scatter(x='Spending Score (1-100)',y='Age',c='DarkBlue')
plt.title('Correlation between Age and Spending Score')
plt.show()
#print(scat1)

printmd("<h3>Correlation between Age and Annual Income\n")
corr2 = df['Age'].corr(df['Annual Income (k$)'])
print(corr2)  

scat2=df.plot.scatter(x='Annual Income (k$)',y='Age',c='DarkBlue')
#print(scat2)
plt.title('Correlation between Age and Annual Income')
plt.show()

printmd("<h3>Correlation between Annual Income and Spending Score \n")
corr3 = df['Spending Score (1-100)'].corr(df['Annual Income (k$)'])

plt.show()
print(corr3)  

scat3=df.plot.scatter(x='Annual Income (k$)',y='Spending Score (1-100)',c='DarkBlue')

plt.show()
#print(scat3)
###########################################################################################################################


printmd("<h2> DIMENSION REDUCTION")

#del dfnew['age_group']
del dfnew['Gender']
del dfnew['CustomerID']

#Performing Standardization
scaler=MinMaxScaler((0,1))
X=scaler.fit_transform(dfnew)

dfnewt=dfnew.copy(deep=True) #Making a copy

#Performing PCA
printmd("<h3> Performing PCA on non-standardized data\n")

pca = PCA(n_components = 2)
pca.fit(dfnew)
pca_X = pca.transform(dfnew)
print(pca_X)

printmd("<h3> Performing PCA on Standardized data\n")
pca = PCA(n_components = 2)
pca.fit(X)
pca_stand = pca.transform(X)
print(pca_stand)

printmd("<h3> PCA on standardized data\n")
plt.scatter(pca_stand[:,0], pca_stand[:,1], s=20, alpha=0.8)
plt.title("Applying PCA on standardized data")
plt.show()

printmd("<h3> PCA on non-standardized data\n")
plt.scatter(pca_X[:,0], pca_X[:,1], s=20, alpha=0.8)
plt.title("Applying PCA on non-standardized data")
plt.show()
```

<h1>Task 2 – Clustering</h1>

```python
#Elbow method
# Using the elbow method to find the optimal number of clusters
#from sklearn.cluster import KMeans
printmd("<h3> Using Elbow method to find the optimal value of k\n")
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(dfnew)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(10,5))
sns.lineplot(range(1, 11), wcss,marker='o',color='red')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.axvline(x=5, color='blue', linestyle='--')
plt.grid(True)
plt.show()

#Using Silhouette method to find out the number of clusters for KMeans
printmd("<h3> Using Silhouette method to find the optimal value of k\n")
sil=[]
kmax=10
for k in range(2,kmax+1):
    kmeans=KMeans(n_clusters=k,random_state=0)
    kmeans.fit(dfnew)
    labels=kmeans.labels_
    sil.append(silhouette_score(dfnew,labels,metric='euclidean'))


plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(2,11),sil,linewidth=2,color='#6495ED',marker='8')
plt.axvline(6, ls="--", c="red")
plt.grid()
plt.title("The Silhouette Method")
plt.xlabel("K Value")
plt.ylabel('Silhouette Score')
plt.show()

#Converting the original dataset to a numpy array
dfnewt=dfnewt.to_numpy()



printmd("<h3>Without PCA or Standardization(KMEANS,K=5)")
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y1 = kmeans.fit_predict(dfnewt)

# Visualising the clusters
plt.figure(figsize=(15,7))
sns.scatterplot(dfnewt[y1 == 0, 0], dfnewt[y1 == 0, 1], color = 'yellow', label = 'Cluster 1',s=50)
sns.scatterplot(dfnewt[y1 == 1, 0], dfnewt[y1 == 1, 1], color = 'blue', label = 'Cluster 2',s=50)
sns.scatterplot(dfnewt[y1 == 2, 0], dfnewt[y1 == 2, 1], color = 'green', label = 'Cluster 3',s=50)
sns.scatterplot(dfnewt[y1 == 3, 0], dfnewt[y1 == 3, 1], color = 'grey', label = 'Cluster 4',s=50)
sns.scatterplot(dfnewt[y1 == 4, 0], dfnewt[y1 == 4, 1], color = 'orange', label = 'Cluster 5',s=50)
sns.scatterplot(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color = 'red', label = 'Centroids',s=300,marker='*')
plt.grid(False)
plt.title('Clusters of customers (Without PCA or Standardization,KMEANS,K=5)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
###########################################################################################################################
printmd("<h3>With PCA (KMEANS,K=5)")
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y2 = kmeans.fit_predict(pca_X)

# Visualising the clusters
plt.figure(figsize=(15,7))
sns.scatterplot(pca_X[y2 == 0, 0], pca_X[y2 == 0, 1], color = 'yellow', label = 'Cluster 1',s=50)
sns.scatterplot(pca_X[y2 == 1, 0], pca_X[y2 == 1, 1], color = 'blue', label = 'Cluster 2',s=50)
sns.scatterplot(pca_X[y2 == 2, 0], pca_X[y2 == 2, 1], color = 'green', label = 'Cluster 3',s=50)
sns.scatterplot(pca_X[y2 == 3, 0], pca_X[y2 == 3, 1], color = 'grey', label = 'Cluster 4',s=50)
sns.scatterplot(pca_X[y2 == 4, 0], pca_X[y2 == 4, 1], color = 'orange', label = 'Cluster 5',s=50)
sns.scatterplot(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color = 'red', label = 'Centroids',s=300,marker='*')
plt.grid(False)
plt.title('Clusters of customers(With PCA (KMEANS,K=5))')
# plt.xlabel('Annual Income (k$)')
# plt.ylabel('Spending Score (1-100)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
###########################################################################################################################
printmd("<h3>With PCA and Standardization (KMEANS,K=5)")
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y3 = kmeans.fit_predict(pca_stand)

# Visualising the clusters
plt.figure(figsize=(15,7))
sns.scatterplot(pca_stand[y3 == 0, 0], pca_stand[y3 == 0, 1], color = 'yellow', label = 'Cluster 1',s=50)
sns.scatterplot(pca_stand[y3 == 1, 0], pca_stand[y3 == 1, 1], color = 'blue', label = 'Cluster 2',s=50)
sns.scatterplot(pca_stand[y3 == 2, 0], pca_stand[y3 == 2, 1], color = 'green', label = 'Cluster 3',s=50)
sns.scatterplot(pca_stand[y3 == 3, 0], pca_stand[y3 == 3, 1], color = 'grey', label = 'Cluster 4',s=50)
sns.scatterplot(pca_stand[y3 == 4, 0], pca_stand[y3 == 4, 1], color = 'orange', label = 'Cluster 5',s=50)
sns.scatterplot(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color = 'red', label = 'Centroids',s=300,marker='*')
plt.grid(False)
plt.title('Clusters of customers (With PCA and Standardization (KMEANS,K=5))')
# plt.xlabel('Annual Income (k$)')
# plt.ylabel('Spending Score (1-100)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()



###########################################################################################################################
printmd("<h3>Without PCA or Standardization(KMEANS,K=6)")
kmeans = KMeans(n_clusters = 6, init = 'k-means++', random_state = 42)
y1 = kmeans.fit_predict(dfnewt)

# Visualising the clusters
plt.figure(figsize=(15,7))
sns.scatterplot(dfnewt[y1 == 0, 0], dfnewt[y1 == 0, 1], color = 'yellow', label = 'Cluster 1',s=50)
sns.scatterplot(dfnewt[y1 == 1, 0], dfnewt[y1 == 1, 1], color = 'blue', label = 'Cluster 2',s=50)
sns.scatterplot(dfnewt[y1 == 2, 0], dfnewt[y1 == 2, 1], color = 'green', label = 'Cluster 3',s=50)
sns.scatterplot(dfnewt[y1 == 3, 0], dfnewt[y1 == 3, 1], color = 'grey', label = 'Cluster 4',s=50)
sns.scatterplot(dfnewt[y1 == 4, 0], dfnewt[y1 == 4, 1], color = 'orange', label = 'Cluster 5',s=50)
sns.scatterplot(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color = 'red', label = 'Centroids',s=300,marker='*')
plt.grid(False)
plt.title('Clusters of customers (Without PCA or Standardization(KMEANS,K=6))')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
###########################################################################################################################
printmd("<h3>With PCA (KMEANS,K=6)")
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 6, init = 'k-means++', random_state = 42)
y2 = kmeans.fit_predict(pca_X)

# Visualising the clusters
plt.figure(figsize=(15,7))
sns.scatterplot(pca_X[y2 == 0, 0], pca_X[y2 == 0, 1], color = 'yellow', label = 'Cluster 1',s=50)
sns.scatterplot(pca_X[y2 == 1, 0], pca_X[y2 == 1, 1], color = 'blue', label = 'Cluster 2',s=50)
sns.scatterplot(pca_X[y2 == 2, 0], pca_X[y2 == 2, 1], color = 'green', label = 'Cluster 3',s=50)
sns.scatterplot(pca_X[y2 == 3, 0], pca_X[y2 == 3, 1], color = 'grey', label = 'Cluster 4',s=50)
sns.scatterplot(pca_X[y2 == 4, 0], pca_X[y2 == 4, 1], color = 'orange', label = 'Cluster 5',s=50)
sns.scatterplot(pca_X[y2 == 5, 0], pca_X[y2 == 5, 1], color = 'cyan', label = 'Cluster 5',s=50)
sns.scatterplot(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color = 'red', label = 'Centroids',s=300,marker='*')
plt.grid(False)
plt.title('Clusters of customers(With PCA (KMEANS,K=6))')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()

###########################################################################################################################
printmd("<h3>With PCA and Standardization (KMEANS,K=6)")
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 6, init = 'k-means++', random_state = 42)
y3 = kmeans.fit_predict(pca_stand)

# Visualising the clusters
plt.figure(figsize=(15,7))
sns.scatterplot(pca_stand[y3 == 0, 0], pca_stand[y3 == 0, 1], color = 'yellow', label = 'Cluster 1',s=50)
sns.scatterplot(pca_stand[y3 == 1, 0], pca_stand[y3 == 1, 1], color = 'blue', label = 'Cluster 2',s=50)
sns.scatterplot(pca_stand[y3 == 2, 0], pca_stand[y3 == 2, 1], color = 'green', label = 'Cluster 3',s=50)
sns.scatterplot(pca_stand[y3 == 3, 0], pca_stand[y3 == 3, 1], color = 'grey', label = 'Cluster 4',s=50)
sns.scatterplot(pca_stand[y3 == 4, 0], pca_stand[y3 == 4, 1], color = 'orange', label = 'Cluster 5',s=50)
sns.scatterplot(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color = 'red', label = 'Centroids',s=300,marker='*')
plt.grid(False)
plt.title('Clusters of customers (With PCA and Standardization (KMEANS,K=6))')
# plt.xlabel('Annual Income (k$)')
# plt.ylabel('Spending Score (1-100)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()


printmd("<h3> It can be seen that k=5 gives better clusters.\n")
printmd("<h3> Calculating the silhouette scores:.\n")

from sklearn.metrics import silhouette_samples, silhouette_score
score1= silhouette_score(dfnewt, y1, metric='euclidean')
# Print the score
print('Silhouette Score for kmeans without PCA or standarization: %.3f' % score1)
score2 = silhouette_score(pca_X, y2, metric='euclidean')
# Print the score
print('Silhouette Score for kmeans with PCA applied data: %.3f' % score2)
score3 = silhouette_score(pca_stand, y3, metric='euclidean')
# Print the score
print('Silhouette Score for kmeans with PCA and standardized applied data: %.3f' % score3)



#DBSCAN
# distances = np.sort(distances, axis=0)
# distances = distances[:,1]
# plt.axhline(y=8.5, color='red', linestyle='--')
# plt.plot(distances)

printmd("<h3>Using Knee method to find suitable value for eps")
from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors(n_neighbors=4)
nbrs = neigh.fit(dfnewt)
distances, indices = nbrs.kneighbors(dfnewt)
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.axhline(y=8.5, color='red', linestyle='--')
plt.plot(distances)
plt.title("The Knee method")


#DBSCAN

###########################################################################################################################
printmd("<h3>Without PCA or Standardization(DBSCAB,eps=8.5)")
db1=DBSCAN(eps=8.5,min_samples=4)
db1.fit(dfnewt)
labels1=db1.fit_predict(dfnewt)
plt.figure(figsize=(15,7))
plt.title("Data points in cluster (Without PCA or Standardization(DBSCAB,eps=8.5))")
plt.scatter(dfnewt[:,0],dfnewt[:,1],c=labels1, cmap='Spectral')
plt.show()

###########################################################################################################################
printmd("<h3>With PCA (DBSCAB,eps=8.5)")
db2=DBSCAN(eps=8.5,min_samples=4)
db2.fit(pca_X)
labels2=db2.fit_predict(pca_X)
plt.figure(figsize=(15,7))
plt.title("Data points in cluster (With PCA (DBSCAB,eps=8.5))")
plt.scatter(pca_X[:,0],pca_X[:,1],c=labels2, cmap='Spectral')
plt.show()

###########################################################################################################################
printmd("<h3>With PCA and Standardization(DBSCAB,eps=8.5)")
db3=DBSCAN(eps=8,min_samples=4)
db3.fit(pca_stand)
labels3=db3.fit_predict(pca_stand)
plt.figure(figsize=(15,7))
plt.title("Data points in cluster(With PCA and Standardization(DBSCAB,eps=8.5))")
plt.scatter(pca_stand[:,0],pca_stand[:,1],c=labels3, cmap='Spectral')
plt.show()
###########################################################################################################################

printmd("<h3> Calculating the silhouette scores:.\n")
score1= silhouette_score(dfnewt, labels1, metric='euclidean')
# Print the score
print('Silhouette Score for dbscan without PCA or standarization: %.3f' % score1)
score2 = silhouette_score(pca_X, labels2, metric='euclidean')
# Print the score
print('Silhouette Score for dbscan with PCA applied data: %.3f' % score2)
#score3 = silhouette_score(pca_stand, labels3, metric='euclidean')
# Print the score
#print('Silhouette Score for dbscan with PCA and standardized applied data: %.3f' % score3)

printmd("<h3> Comparing the performances for K-means and DBSCAN\n")
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
algorithm = ['K-means', 'DBSCAN']
score = [0.536,0.214]
ax.bar(algorithm,score)
plt.title('Comaprison of performances')
plt.xlabel('Clustering Algorithm')
plt.ylabel('Silhouette score ')
plt.show()
```
